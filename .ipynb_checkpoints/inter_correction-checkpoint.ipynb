```python
#COUNTS = Counter(all_word)
alphabet = 'abcdefghijklmnopqrstuvwxyz0123456789.,/=+-*?!'

def correct_spelling(word,COUNTS):
    "Find the best spelling correction for this word."
    # Prefer edit distance 0, then 1, then 2; otherwise default to word itself.
    candidates = (known(edits0(word),COUNTS) or 
                  known(edits1(word),COUNTS) or 
                  known(edits2(word),COUNTS) or 
                  #known(edits3(word),COUNTS) or 
                  [word])
    return max(candidates, key=COUNTS.get)

def known(words,COUNTS):
    "Return the subset of words that are actually in the dictionary."
    return {w for w in words if w in COUNTS}

def edits0(word): 
    "Return all strings that are zero edits away from word (i.e., just word itself)."
    return {word}

def edits1(word):
    "Return all strings that are one edit away from this word."
    pairs      = splits(word)
    deletes    = [a+b[1:]           for (a, b) in pairs if b]
    transposes = [a+b[1]+b[0]+b[2:] for (a, b) in pairs if len(b) > 1]
    replaces   = [a+c+b[1:]         for (a, b) in pairs for c in alphabet if b]
    inserts    = [a+c+b             for (a, b) in pairs for c in alphabet]
    return set(deletes + transposes + replaces + inserts)

def edits2(word):
    "Return all strings that are two edits away from this word."
    return {e2 for e1 in edits1(word) for e2 in edits1(e1)}

def edits3(word):
    "Return all strings that are two edits away from this word."
    return {e3 for e1 in edits1(word) for e2 in edits1(e1) for e3 in edits1(e2) }

def splits(word):
    "Return a list of all possible (first, rest) pairs that comprise word."
    return [(word[:i], word[i:]) 
            for i in range(len(word)+1)]
```

```python
def split_mul_word(temp_word,vocabulary):
    word_list=[]
    #num_list=[]
    if len(re.findall('[a-z]+', temp_word))>0:
        '''
        if len(re.findall('[1-9]+', temp_word))>0:
            en_word=re.findall('[a-z]+', temp_word)
            word_list=word_list+en_word
            #num_word=re.findall('[1-9]+', temp_word)
            #word_list=word_list+en_word+num_word
        '''
        if '.' in temp_word or ',' in temp_word or '/' in temp_word or '!' in temp_word or len(re.findall('[1-9]+', temp_word))>0:
            en_word=re.findall('[a-z]+', temp_word)
            word_list=word_list+en_word
        elif len(temp_word)>=6:
            flag=0
            for i in range(1,len(temp_word)-1):
                if reform_word(temp_word[:i]) in vocabulary and reform_word(temp_word[i:]) in vocabulary:
                    word_list=word_list+[temp_word[:i]]+[temp_word[i:]]
                    flag=1
                    break
            if flag==0:
                word_list=[temp_word]
            
            
        
    #elif len(re.findall('[1-9]+', temp_word))>1:
    
    else:
        word_list=[temp_word]
                               
    return word_list
```

```python
def correct_words_once(word_list):
    final_word_list=[]
    for temp_word in word_list:
        if temp_word in VOCABULARY:
            final_word_list=final_word_list+[temp_word]
        elif len(re.findall('[a-z]+', temp_word))==1:
            #print(temp_word+':   need correction')
            try:
                temp_word_freq=word_dict[temp_word]
            except:
                temp_word_freq=0
            more_freq_words=[temp[0] for temp in sort_word_dict if temp[1]>temp_word_freq]
            filtered_words=[temp_all_word for temp_all_word in all_word if temp_all_word in more_freq_words]
            temp_counter=Counter(filtered_words)
            correct_word=correct_spelling(temp_word,temp_counter)
            final_word_list=final_word_list+[correct_word]
        else:
            final_word_list=final_word_list+[temp_word]
    return final_word_list
```

```python
def transfer_answer(temp_answer,freq_voc, rare_voc):
    temp_answer=str(temp_answer).lower()
    word_list=[]
    num_list=[]
    symb_list=[]
    symb_range=['/','+','*','-','=']
    temp_words=re.split(' ', temp_answer)
    #temp_words=re.findall('[a-z]+', temp_answer) 
    
    for temp_word in temp_words:
        if temp_word == '':
            continue
        process_word=reform_word(temp_word)
        #process_word=porter.stem(process_word)
        #process_word=lemmatizer.lemmatize(process_word)
        #if not(process_word in freq_voc) and not(process_word in VOCABULARY):
        if not(process_word in VOCABULARY):
            #print(process_word+':   need correction')
            temp_list=split_mul_word(temp_word,freq_voc)
            if len(temp_list)>1:
                for i in range(len(temp_list)):
                    temp_process_word=reform_word(temp_list[i])
                    word_list=word_list+[temp_process_word]
            else:
                while( len(temp_word)>1 and temp_word[-1] in ['.',',','!','?'] ):
                    temp_word=temp_word[:-1]
                word_list=word_list+[temp_word]
                
        else:
            word_list=word_list+[process_word]
            
    num_list=re.findall('[1-9]+', temp_answer)
    for temp_charac in temp_answer:
        if temp_charac in symb_range:
            symb_list=symb_list+[temp_charac]
            
    '''
    final_word_list=[]
    for temp_word in word_list:
        if temp_word in VOCABULARY:
            final_word_list=final_word_list+[temp_word]
        elif len(re.findall('[a-z]+', temp_word))==1:
            print(temp_word+':   need correction')
            try:
                temp_word_freq=word_dict[temp_word]
            except:
                temp_word_freq=0
            more_freq_words=[temp[0] for temp in sort_word_dict if temp[1]>temp_word_freq]
            filtered_words=[temp_all_word for temp_all_word in all_word if temp_all_word in more_freq_words]
            temp_counter=Counter(filtered_words)
            correct_word=correct_spelling(temp_word,temp_counter)
            final_word_list=final_word_list+[correct_word]
        else:
            final_word_list=final_word_list+[temp_word]
    '''
    last_final_word_list=word_list
    while(True):
        temp_final_word_list=correct_words_once(last_final_word_list)
        if temp_final_word_list==last_final_word_list:
            break
        last_final_word_list=temp_final_word_list.copy()
        
        
        
    final_word_list=temp_final_word_list
    

    
    return final_word_list,num_list,symb_list
```
